## Упражнение 9.3
1. С помощью библиотеки OpenCV считайте изображение:

```python
IMG_PATH_3 = "data\\Task_2\\3.jpg"
image_3 = cv2.imread(IMG_PATH_3)
```

2. Сожмите изображение до ширины `720`, соблюдая пропорции исходного изображения. Важно:коэффициент отношения сторон при вычислениях не округляйте. Если высота не окажется целой, округлите значение до целого в меньшую сторону. Используйте `cv2.resize()` с параметром `interpolation = cv2.INTER_AREA`.

Введите коэффициент отношения исходной ширины изображения к его высоте (Ответ введите с точностью до тысячных, однако для вычисления высоты используйте точное значние):

```python
h, w, c = image_3.shape
resized_image = cv2.resize(image_3, (720, int(h / w * 720)), interpolation = cv2.INTER_AREA)
plt.imshow(resized_image)
print("Коэффициент: %.3f" % (w / h))
print("Новая h:", resized_image.shape[0])
```

3. Сохраните сжатое изображение в палитре RGB.

Введите размер сжатого изображения в байтах:

```python
OUTPUT_PATH = "data\\Task_2\\output.jpg"
cv2.imwrite(OUTPUT_PATH, resized_image)
os.path.getsize(OUTPUT_PATH) 
```
